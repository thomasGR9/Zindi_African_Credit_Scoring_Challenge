```python
data_tryouts['disbursement_date'] = pd.to_datetime(data_tryouts['disbursement_date'])
```

```python
data_tryouts['due_date'] = pd.to_datetime(data_tryouts['due_date'])
```

```python
from datetime import datetime
```

```python
overall_weights_list = []


for _, row in data_tryouts.iterrows():
    sample_weights_list=[]
    disbursement_date = row['disbursement_date']
    due_date = row['due_date']
    disbursement_year = disbursement_date.year
    due_year = due_date.year

    if disbursement_year == due_year:
        sample_weights_list.append(1)

    else:
        years_between = due_year-disbursement_year
        days_of_loan = row['duration']
        # Calculate days until the end of the year for disbursement_date
        end_of_year = datetime(disbursement_year, 12, 31)
        days_until_end_of_year = (end_of_year - disbursement_date).days
        sample_weights_list.append(days_until_end_of_year/days_of_loan)
        for i in range(years_between-1):
            sample_weights_list.append(365/days_of_loan)

        # Calculate days passed since the start of the year for due_date
        start_of_year = datetime(due_year, 1, 1)
        days_passed_from_start_of_year = (due_date - start_of_year).days
        sample_weights_list.append(days_passed_from_start_of_year/days_of_loan)
    overall_weights_list.append(sample_weights_list)
```

```python
for index, row in data_tryouts.iterrows():
    start_year_row = row['disbursement_date'].year
    row_country = row['country_id']
    for indicator in economic_index_data['Indicator'].unique():
        if indicator not in data_tryouts.columns:
            data_tryouts[indicator] = 0
        value=0
        for i in range(len(overall_weights_list[index])):
            check_year=start_year_row+i
            if (check_year>2023):
                check_year=2023
            value += (overall_weights_list[index][i]*economic_index_data[(economic_index_data['Indicator']==indicator) & (economic_index_data['Country']==row_country)][f"YR{check_year}"].values[0])
        data_tryouts.at[index, indicator] = value
        
    
```

```python
data_tryouts['repay_pct_lender'] = data_tryouts['repay_pct_lender'].fillna(0)
```

```python
data_tryouts['repay_pct_overall'] = data_tryouts['Total_Amount_to_Repay'] / data_tryouts['Total_Amount']
```

```python
data_tryouts['repay_amount_overall'] = data_tryouts['Total_Amount_to_Repay'] - data_tryouts['Total_Amount']
```

```python
data_tryouts['repay_amount_lender'] = data_tryouts['Lender_portion_to_be_repaid'] - data_tryouts['Amount_Funded_By_Lender']
```

```python
data_tryouts['amount_per_day'] = data_tryouts['repay_amount_lender'] / data_tryouts['duration']
```

```python
total_borrowed_money_list = []
for customer in data_tryouts['customer_id'].values:
    total_borrowed_money  = 0
    for loan_id in data_tryouts[(data_tryouts['customer_id']==customer)]['tbl_loan_id'].unique():
        loan = data_tryouts[(data_tryouts['customer_id']==customer) & (data_tryouts['tbl_loan_id']==loan_id)]['Total_Amount'].values
        total_borrowed_money += loan[0]
    total_borrowed_money_list.append(total_borrowed_money)

        
    
        
        
        
    
```

```python
data_tryouts['total_money_borrowed'] = total_borrowed_money_list
```

```python
number_of_loans_per_customer_list = []
sum_of_lender_loans_per_customer_list = []
for customer_id in data_tryouts['customer_id'].values:
    number_of_loans_per_customer_list.append(data_tryouts[data_tryouts['customer_id']==customer_id].shape[0])
    sum_of_lender_loans_per_customer_list.append(data_tryouts[data_tryouts['customer_id']==customer_id]['Amount_Funded_By_Lender'].values.sum())
```

```python
data_tryouts['total_number_of_loans']=number_of_loans_per_customer_list
```

```python
data_tryouts['sum_of_lender_loans'] = sum_of_lender_loans_per_customer_list
```

```python
data_tryouts['mean_amount_per_loan'] = data_tryouts['sum_of_lender_loans'] / data_tryouts['total_number_of_loans']
```

```python
data_tryouts['usual_borrowed_ratio'] = data_tryouts['Amount_Funded_By_Lender'] / data_tryouts['mean_amount_per_loan']
```
