```python
x_df = data.drop('target', axis=1)
y_arr = data['target'].to_numpy()
```

```python
x_df
```

```python
unbalance_ratio = (y_arr==1).sum() / (y_arr==0).sum() 
ratio_of_pos = (y_arr==1).sum() / (y_arr.shape[0])
number_of_val_pos = int((x_df.shape[0] / 10)*ratio_of_pos)
number_of_val_neg = int((x_df.shape[0] / 10) - number_of_val_pos)
indexes_of_pos = np.where(y_arr==1)
indexes_of_neg = np.where(y_arr==0)
```

```python
x_neg_df = x_df.loc[indexes_of_neg]
```

```python
x_neg_val_df = x_neg_df.sample(n=number_of_val_neg, random_state=42)
```

```python
y_neg_val = y_arr[x_neg_val_df.index]
x_neg_val_arr = x_neg_val_df.to_numpy()
```

```python
x_one_arr = x_df.loc[indexes_of_pos].to_numpy()
```

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=number_of_val_pos, random_state=42)
x_dist = km.fit_transform(X=x_one_arr[:, 3:])
most_representative_pos = np.argmin(x_dist, axis=0)
x_pos_repr_val = x_one_arr[most_representative_pos]
```

```python
y_pos_val = np.full(shape=(x_pos_repr_val.shape[0]),fill_value=1)
```

```python
x_val_stacked = np.vstack((x_pos_repr_val, x_neg_val_arr))  
y_val_stacked = np.hstack((y_pos_val, y_neg_val))        
indices = np.arange(x_val_stacked.shape[0])  
np.random.shuffle(indices)       
x_val_ID = x_val_stacked[indices]
x_val=x_val_ID[:,1:]
y_val = y_val_stacked[indices]
```

```python
x_val_ID
```

```python
indexes_to_exclude = set(x_neg_val_df.index).union(x_df.loc[indexes_of_pos].iloc[most_representative_pos].index)  # Combine both index sets
x_train_df = x_df.drop(index=indexes_to_exclude)
x_train = x_train_df.drop(['ID'], axis=1).to_numpy()
y_train = y_arr[x_train_df.index]
```

```python
x_num_train = x_train_df.drop(['ID', 'loan_type', 'lender_id'], axis=1).to_numpy()
x_cat_train = x_train_df[['ID', 'lender_id','loan_type']].to_numpy()
```

```python
from imblearn.over_sampling import SMOTENC
```

```python
cat_columns = ['loan_type', 'lender_id']
sm = SMOTENC(random_state=42, categorical_features=cat_columns)
```

```python
x_pca_train = pca.fit_transform(X=x_train_scaled_df.drop(cat_columns, axis=1))
```
